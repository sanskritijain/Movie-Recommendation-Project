# Creating Movie Recommendation

```python
moviemat = df.pivot_table(index ='user_id ',columns = 'title',values ='ratings')
moviemat
```

```python
starwar_userRatings = moviemat['Star Wars (1977)']
```

```python
similar_to_starwars = moviemat.corrwith(starwar_userRatings)


```

```python
corr_starwars = pd.DataFrame(similar_to_starwars ,columns =['Correlation'])
corr_starwars.head()
```

```python
corr_starwars.dropna(inplace = True)
corr_starwars
```

```python
corr_starwars.sort_values('Correlation' , ascending = False)
```

```python
corr_starwars = corr_starwars.join(rating['no of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['no of ratings'] >100].sort_values('Correlation' ,ascending = False)
```

# Predict Function 

```python
def predict(movie_name) : 
    movie_userRatings = moviemat[movie_name]
    similar_to_movie = moviemat.corrwith(movie_userRatings)
    
    corr_movie = pd.DataFrame(similar_to_movie ,columns =['Correlation'])
    corr_movie.dropna(inplace = True)
    
    corr_movie = corr_movie.join(rating['no of ratings'])
    predictions= corr_movie[corr_movie['no of ratings'] >100].sort_values('Correlation' ,ascending = False)
    
    return predictions
    
    
```

```python
predictions = predict('Titanic (1997)')
predictions
```
